```python
from A3 import iter

t = np.linspace(-2,2,100)
X, Y = np.meshgrid(t,t)
Z = X + 1j * Y

R = np.zeros(Z.shape)
for i in range(len(X)):
    for h in range(len(Y)):
        R[i,h] = iter(0, Z[i,h], 100)
print(R)
```

```python
from scipy.integrate import solve_ivp

def atmosphere(t, x, sigma, r ,b):
    return [-sigma*(x[0]-x[1]), r*x[0]-x[1]-x[0]*x[2], -b*x[2]+x[0]*x[1]]

arng = (10,28,8/3)

T_max = 60
t_vals = np.linspace(0, T_max, 1000)
sol = solve_ivp(fun = lambda t, x:atmosphere(t,x,*arng),\
                t_span = (0, T_max),\
                y0 = [0,1,0],\
                t_eval = t_vals)

plt.plot(sol.t, sol.y[0], label = '$X$')
plt.plot(sol.t, sol.y[1], label = '$Y$')
plt.plot(sol.t, sol.y[2], label = '$Z$')
plt.legend()
```

```python
plt.figure(figsize=(12,4))
plt.plot(sol.t/0.01, sol.y[1], label = '$Y$')
```

```python
T_max_re = 30
t_vals_re = np.linspace(0, T_max_re, 3000)
sol_re = solve_ivp(fun = lambda t, x:atmosphere(t,x,*arng),\
                   t_span = (0, T_max_re),\
                   y0 = [0,1,0],\
                   t_eval = t_vals_re)

fig, ax = plt.subplots(nrows=3,ncols=1, figsize = (15,5))
for i in range(0,3):
    ax[i].plot(sol_re.t/0.01, sol_re.y[1], label = '$X$')
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].set_xlim(i*1000, (i+1)*1000)

```

```python
import unittest

def Area(r):
    """Return the area of a circle with radius 'r'
    
    input: r (float) radius of the circle
    return: area (float) area of the circle
    """
    return np.pi * r**2

class TestIsDivThree(unittest.TestCase):

    def test_3(self):
        self.assertEqual(Area(3),np.pi*3**2)
    def test_4(self):
        self.assertEqual(Area(4),np.pi*4**2)

unittest.main(argv=[''], verbosity=2, exit=False)
```
